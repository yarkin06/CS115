Generate two arrays(x and y) each containing 100 random points that fall in the square of size 6. The points will have x coordinates in the interval [-3,3] and y coordinates in the interval [-3,3].

```python
from numpy import *
```

```python
x = 6*random.rand(100)-3
y = 6*random.rand(100)-3
print(x)
print(y)
```

Find the number of points which are inside the circle centered at the origin and having radius 3, as shown in the figure below.


```python
count = 0
x_in = []
y_in = []
for i in range(100):
    if x[i]**2 + y[i]**2 <= 9:
        count += 1
        x_in.append(x[i])
        y_in.append(y[i])
        
print(count)
```

Join the arrays containing the x and y coordinates, using the np.stack() function, into a new 2D array with n rows (where n represents the number of x and y coordinates inside the circle) and 2 columns.  To do this you may use the stack function.  Stack takes the (arrays), the axis=1(to concatenate columns).

```python
print(stack((x_in,y_in),1))
```
